**Dish Plate Count**

```python
class PlateCount:
  def enterValue(self):
    num=int(input('Enter: '))
    return num

  def DetailOfOrder(self):
    print('*** Dish Detail You Order *** \t')
    print('-'*70)
    if numMain == 1:
      print('Dish Name:   \t',VegItemList().nameOfItem())
      print('Dish Price:  \t',VegItemList().priceOfItem())
      print('-'*70)
    elif numMain == 2:
      print('Dish Name:   \t',NonVegItemList().nameOfItem())
      print('Dish Price:  \t',NonVegItemList().priceOfItem())
      print('-'*70)
    elif numMain == 3:
      print('Dish Name:   \t',DesertItemList().nameOfItem())
      print('Dish Price:  \t',DesertItemList().priceOfItem())
      print('-'*70)

  def dishPlateCount(self):
    print('Please Enter number of plates you want to order:')
    global get_numplates
    get_numplates=self.enterValue()
    return get_numplates
```

**Billing Section**

```python
amount = 0
class BillingSection:

  def billingProcess(self):
    get_dishPlateCount=PlateCount().dishPlateCount()
    PlateCount().DetailOfOrder()
    global total
    if numMain == 1:
      total =  VegItemList().priceOfItem() * get_dishPlateCount
    elif numMain == 2:
      total =  NonVegItemList().priceOfItem() * get_dishPlateCount
    elif numMain == 3:
      total =  DesertItemList().priceOfItem() * get_dishPlateCount




  def totalBill(self):
    print('\n')
    print('-'*70)
    print('*** Total Bill Price ***')
    global amount
    amount = amount + total
    print('-'*70)
    print('Total Amount: ',amount)
    print('-'*70)

```

**Payement Method**

```python
class Payement:
  def enterValue(self):
    num=int(input('Enter: '))
    return num

  def cash(self):
    print(' Cash Transaction Done')

  def card(self):
    print('Card Transaction Done')

  def choosePayementMethod(self):
    print('''Choose Payement Option
          1. Cash
          2. Card''')
    cashNum = self.enterValue()
    while True:
      if cashNum == 1:
        self.cash()
        break
      elif cashNum == 2:
        self.card()
        break
      else:
        print('Please Select from above method')
        cashNum = self.enterValue()
```
